```python
# Create results summary
import pandas as pd
import json

# Collect final metrics from training logs
models = ['baseline', 'se', 'cbam', 'hybrid']
results = []

for model in models:
    try:
        with open(f'outputs/{model}/training_log.json', 'r') as f:
            log = json.load(f)
            results.append({
                'Model': model.upper(),
                'Final WT Dice': log.get('final_wt_dice', 'N/A'),
                'Final TC Dice': log.get('final_tc_dice', 'N/A'),
                'Final WT HD95': log.get('final_wt_hd95', 'N/A'),
                'Final TC HD95': log.get('final_tc_hd95', 'N/A')
            })
    except:
        results.append({'Model': model.upper(), 'Status': 'Training incomplete'})

# Display results table
df = pd.DataFrame(results)
print("\n📊 Final Results Summary:")
print(df.to_string(index=False))

# Save results
df.to_csv('results/model_comparison.csv', index=False)
print("\n💾 Results saved to results/model_comparison.csv")
```
